```python
# Explain query
spark.sql("SELECT * FROM events_tbl WHERE event_type='purchase'").explain(True)
```

```python

spark.sql("""
  CREATE TABLE events_part
  USING DELTA
  PARTITIONED BY ( event_type)
  AS SELECT * FROM events_tbl
""")
```

```python

spark.sql("OPTIMIZE events_part ZORDER BY (user_id, product_id)")
```
