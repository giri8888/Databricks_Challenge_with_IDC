```python
events = spark.read.option("header","true").option("inferSchema","true").csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv")
```

```python
display(events)
```

```python
events.printSchema()
```

```python
events.write.format("delta").mode("overwrite").save("/Volumes/workspace/ecommerce/ecommerce_data/events")
```

```python
events.write.format("delta").saveAsTable("events_tbl")
events_tbl = spark.read.table("events_tbl")
display(events_tbl)
```
