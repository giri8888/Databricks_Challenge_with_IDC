```python
%sql
-- Revenue with 7-day moving average
WITH daily AS (
  SELECT event_time, SUM(price) as rev
  FROM ecommerce.gold.products GROUP BY event_time
)
SELECT event_time, rev,
  AVG(rev) OVER (ORDER BY event_time ROWS BETWEEN 6 PRECEDING AND CURRENT ROW) as ma7
FROM daily;
```

```python
%sql
-- Conversion funnel
SELECT category_code,
  SUM(views) as views,
  SUM(purchases) as purchases,
  ROUND(SUM(purchases)*100.0/SUM(views), 2) as conversion_rate
FROM ecommerce.gold.products
GROUP BY category_code;
```

```python
%sql
-- Customer tiers
SELECT
  CASE WHEN cnt >= 10 THEN 'VIP'
       WHEN cnt >= 5 THEN 'Loyal'
       ELSE 'Regular' END as tier,
  COUNT(*) as customers,
  AVG(total_spent) as avg_ltv
FROM (SELECT user_id, COUNT(*) cnt, SUM(price) total_spent
      FROM ecommerce.silver.events WHERE event_type='purchase' GROUP BY user_id)
GROUP BY tier;
```
