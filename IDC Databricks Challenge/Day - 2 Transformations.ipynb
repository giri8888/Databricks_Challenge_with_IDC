### Load Data

```python
events = spark.read.option("header","true").option("inferSchema","true").csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv")
```

```python
from pyspark.sql import functions as F
display(events.select(F.col("event_type"),F.col("product_id"),F.col("price")).limit(10))
```

```python
display(events.filter(F.col("price")>100).count())
```

```python
display(events.groupBy("event_type").count())
```

```python
display(events.groupBy("brand").count().orderBy("count", ascending=False).limit(5))
```
